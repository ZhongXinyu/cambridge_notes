If you're flipping a fair coin, what is probability you'll see HTH before TTH? Expected number of flips to see each?

```python
import random
```

```python
simulation = 10000


def game():
    ### let H = 1, T = 0
    chain = []
    for i in range(0,3):
        chain += [random.randint(0, 1)]
    game_end = 0
    while game_end == 0:
        if chain[-3:] == [1,0,1]: ### HTH
            print (chain)
            return (0, len(chain))
        # if chain[-3:] == [0,0,1]: ### TTH
        #     print (chain)
        #     return (1, len(chain))
        chain += [random.randint(0, 1)]

win_count = 0
step = 0
for i in range (simulation):
    win_count += game()[0]
    step += game()[1]

print ("winning rate: ", win_count/simulation)
print ("average step: ", step/simulation)


print(5/8)
print(31/4)

```
