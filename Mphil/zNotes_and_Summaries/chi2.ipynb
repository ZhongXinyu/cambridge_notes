```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

for df in range (1,6):

    # Generate x values
    x = np.linspace(0, 20, 1000)
    # Calculate the chi-square PDF
    pdf = chi2.pdf(x, df)

    # Calculate the expected value
    expected_value = chi2.mean(df)

    # Plot the PDF
    plt.plot(x, pdf, label=f'Chi-square PDF (df={df}), with expected value {expected_value}')

plt.title('Chi-square Distribution PDF')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
