Plot the data from data file $\mathrm{q} 1$.csv. Perform $k$-means for a range of values of $K$. Evaluate the silhouette score in each case and use this to choose an optimal number of clusters (note: we will cover silhouette scores on Mon 27 th Nov).

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
### read data from csv file
data = np.genfromtxt("q1.csv", delimiter=',')
```

```python
print (data.shape)
```

```python
sns.scatterplot(x=data[:,0], y=data[:,1])
plt.show()
```

```python
# iterate from 1 to 8
for K in range (2, 9):
    # use k-means to cluster the data
    kmeans = KMeans(n_clusters=K, random_state=0, n_init = "auto").fit(data)
    # sns.scatterplot(x=data[:,0], y=data[:,1], hue=kmeans.labels_)
    # plt.show()
    s_score = silhouette_score(data, kmeans.labels_)
    print ("K = ", K, " silhouette_score = ", s_score)
```
