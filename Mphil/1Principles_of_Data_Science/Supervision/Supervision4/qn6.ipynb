A Gaussian Mixture Model question. Take a look at the data in datasets/gmm.npy.
This is an array of two dimensional data which is generated from some number of
multivariate normal distributions (with different means and covariances). Use a Gaus-
sian Mixture Model approach to determine how many separate underlying multivariate
normal distributions you think there are, and estimate the means and covariances of
those underlying distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
with open('gmm.npy', 'rb') as f:
    data = np.load(f)
print(data)
```

```python
# Plot the data
plt.figure(figsize=(10, 10))
plt.scatter(data[:, 0], data[:, 1])
plt.title('Scatter plot of the data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components = 4, random_state = 42)
gmm.fit(data)

labels = gmm.predict(data)
print (gmm.get_params())
print (gmm.means_)
print (gmm.covariances_)
```

```python
# plot the data
plt.figure(figsize=(10, 10))
sns.scatterplot(x = data[:, 0], y= data[:, 1], c=labels)
plt.scatter(gmm.means_[:, 0], gmm.means_[:, 1], color='red', marker='X', s=200, lw=2, label='Cluster Centers')
plt.title('Scatter plot of the data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
