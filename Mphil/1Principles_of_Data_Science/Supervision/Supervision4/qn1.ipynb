1. Familiarise yourself with bootstrapping and jackknife resampling. By generating a
small data sample from a normal distribution, demonstrate that the bias in the sample
variance (without the Bessel correction) can be obtained by a resampling.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate a dataset of from a normal distribution
Nobs = 10000
dset = np.random.normal(size = Nobs)

# Get the mean and standard deviation of the dataset
d_mean = np.mean(dset)
d_std = np.std(dset)
```

```python
# generate some bootstrap samples
Nboot = 5000
bsamps = np.random.choice(dset, size = (Nboot, Nobs), replace = True)

print (bsamps.shape)
```

```python
# Get the mean of each bootstrp sample
b_means = [np.mean(s) for s in bsamps]
b_stds = [np.std(s) for s in bsamps]
print (b_means)
```

```python
# Find the Bias of the sample mean
bias = np.mean(b_means) - d_mean
print (bias)
```

```python
# Find the bias of the sample variance
bias = np.mean(b_stds) - d_std
print (bias)
```

```python
jsamples = np.array([np.delete(dset, i) for i in range (Nobs)])
print (jsamples.shape)
```

```python
# Get the mean and stds of each jackknife sample
j_means = [np.mean(s) for s in jsamples]
j_stds = [np.std(s) for s in jsamples]
```

```python
# Find the bias of the sample variance
bias = (Nobs-1)*(np.mean(j_stds) - d_std)
print (bias)

# should this value be the same as the one for the bootstrap?
```
