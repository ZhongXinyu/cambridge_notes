```python
from random import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
size =  10000
mean = [1,4]
cov = [[9,3],[3,4]]

data = np.random.multivariate_normal(mean, cov, size=size)
df = pd.DataFrame(data, columns=["x", "y"])
```

```python
sns.jointplot(x="x", y="y", data=df)
plt.xlabel('x')
plt.ylabel('y')
plt.plot()
```

```python
function = lambda x,y : x+y
c = 5
dc = 0.1
condition = lambda x,y,c,dc : function(x,y) < c+dc and function(x,y) > c-dc

condition (1,2, c, dc)
```

```python
df["condition"] = df.apply(lambda df: condition(df.x, df.y, c, dc), axis=1)
```
