#### (a) Make some pair plots of the distributions in the dataset and try to become a bit familiar with what the data might be showing.

```python
import pickle
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('ps1.pkl', 'rb') as f:
    data = pickle.load(f)
```

```python
data 
```

#### (a) Make some pair plots of the distributions in the dataset and try to become a bit familiar with what the data might be showing.

```python
sns.pairplot(data, vars=["vx", "vy", "vz", "R2", "p"])
plt.show()
```

#### (b) Make a publication quality plot which shows the two-dimensional distribution between the variables vx and vz.

```python
# sns.pairplot(data, vars=["vx", "vz"])
sns.displot(data, x="vx", y="vy", cbar=True)
plt.title(r"Two dimentional distribution between varables $v_x$ and $v_z$", loc= "center")
plt.xlabel(r"$v_x$")
plt.ylabel(r"$v_z$")
plt.show()
```

#### (c) Compute the correlation and covariance matrices between all variables in the dataset

```python
data[["vx","vy","vz"]].cov()
```

```python
data[["vx","vy","vz","R2","p"]].corr()
```

```python
data["new_vx"]=data["vx"].apply(lambda x: x**2 - 1)
sns.pairplot(data, vars=["vx", "new_vx", "R2"])
plt.show()
```
