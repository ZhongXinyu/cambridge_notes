#### 3. The Monty-Hall problem with N doors. In the lectures we saw that in the Monty Hall game show problem with 3 doors it is always advantageous to switch. Sticking on the initial door has a win outcome with p = 1/3, whereas switching has p = 2/3.

```python
import random
import numpy as np
import matplotlib.pyplot as plt 
```

```python
NumberOfTrials = 100000
NumberOfDoors = 3
WinCount_Switch = 0
WinCount_NotSwitch = 0
WinRate_Switch = np.zeros([0])
WinRate_NotSwitch = np.zeros([0])

for i in range (1, NumberOfTrials+1):
    Choice = random.randint(1,NumberOfDoors)
    Prize = random.randint(1,NumberOfDoors)
    if Choice == Prize:
        WinCount_NotSwitch += 1
    else:  
        WinCount_Switch += 1
    WinRate_Switch = np.append (WinRate_Switch, [WinCount_Switch/i])
    WinRate_NotSwitch = np.append (WinRate_NotSwitch, [WinCount_NotSwitch/i])

print ("probability of winning if you switch: ", WinCount_Switch/NumberOfTrials)
print ("probability of winning if you don't switch: ", WinCount_NotSwitch/NumberOfTrials)
```

```python

plt.plot(WinRate_Switch,label = "Switch")
plt.plot(WinRate_NotSwitch,label = "Not Switch")
plt.legend()
plt.show()
```

#### (d) Show that if there are N doors and the host opens p of them that the probability of winning if you switch is $\dfrac{1}{N} \dfrac{N-1}{N-p-1}$ You can always go back to your simulation to check this

Let event $A$ be choosing the door wrong initially, $P(A) = \frac{N-1}{N}$.\
Let event $B$ be choosing the right door after switching.\
Given if you pick the first door worng, the probability of winning after switch is $P(B|A) = \frac{1}{N-p-1}$, as there is 1 right door to choose out of $N-p-1$.\
Given if you pick the first door right, the probabiliry of winning after switch is $P(B|A') = 0$.\
Therefore, the probability of winning if switch $P(B) = P(B|A)P(A) + P(B|A')P(A') = \frac{N-1}{N(N-p-1)}$

#### (e) An extension that you may decide to give up on. Show that if there are N boxes, containing k prizes and the host opens p of them, then the probability of winning if you switch is $\dfrac{k}{N} \dfrac{N-1}{N-p-1}$. You can always go back to your simulation to check this

Let event $A$ be choosing the door wrong initially, $P(A) = \frac{N-k}{N}$.
$P(A') = \frac{k}{N}$\
Let event $B$ be choosing the right door after switching.
After choosing the wrong door $P(B|A) = \frac{k}{N-p-1}$, as there is k right door to choose out of N-p-1.\
After choosing the right door $P(B|A') = \frac{k-1}{N-p-1}$, as there is n-k right door to choose out of N-p-1.\
Therefore, the probability of winning if switch = $P(B) = P(B|A)P(A) + P(B|A')P(A')= \frac{k(N-1)}{N(N-p-1)}$
