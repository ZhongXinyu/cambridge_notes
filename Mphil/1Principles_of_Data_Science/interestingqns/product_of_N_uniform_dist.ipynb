```python
# This simulation discuss the distribution of T, where T is the the product of n independent random variables, each of which is uniformly distributed on [0,1].
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
size = 100
```

```python
from random import random


# generate a random number between 0 and 1
uniform_dist = [random() for i in range(size)]

# plot the distribution of Uniform Distribution
sns.histplot(uniform_dist)
plt.title('Uniform Distribution')
plt.xlabel('Random Number')
plt.ylabel('Frequency')
plt.show()
```

```python
# generate T 
T = [np.prod([random() for _ in range(size)]) for __ in range(size)]

print (T)

# plot the distribution of T
sns.histplot(T)
plt.title('Distribution of T')
plt.xlabel('Random Number')
plt.ylabel('Frequency')
plt.show()
```
