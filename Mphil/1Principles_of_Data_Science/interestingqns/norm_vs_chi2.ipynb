```python
from scipy.stats import norm, chi2
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plot the pdf of a chi2 distribution, k =1 
k = 1 
x = np.linspace(0, 10, 100)
plt.plot(x, chi2.pdf(x, k), label='chi2 pdf')

print (chi2.cdf(1, k))
print (chi2.cdf(4, k))
print (chi2.cdf(9, k))
```
