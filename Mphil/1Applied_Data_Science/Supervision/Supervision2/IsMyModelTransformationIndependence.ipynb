```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt

# Set the desired correlation
desired_correlation = 0.8

# Set the means and standard deviations for x and y
mean_x = 0
std_dev_x = 1
mean_y = 0
std_dev_y = 2

cov_matrix = [[std_dev_x ** 2, desired_correlation * std_dev_x * std_dev_y],[std_dev_x * std_dev_y * desired_correlation, std_dev_y ** 2]]

print (cov_matrix)
# Generate a large number of random samples
n_samples = 1000

# Generate random x values with a normal distribution
data = np.random.multivariate_normal([mean_x, mean_y], cov_matrix, n_samples)

df = pd.DataFrame(data, columns=['x', 'y'])
sns.scatterplot(x="x", y="y", data=df)
plt.show()
```

```python
import math 

theta = 0.24

df["x_prime"] = math.cos(theta) * df["x"] + math.sin(theta) * df["y"]
df["y_prime"] = -math.sin(theta) * df["x"] + math.cos(theta) * df["y"]

sns.scatterplot(x="x_prime", y="y_prime", data=df)
plt.show()
```

```python
print (np.corrcoef(df["x"], df["y"]))
print (np.cov(df["x"], df["y"]))
print (np.corrcoef(df["x_prime"], df["y_prime"]))
print (np.cov(df["x_prime"], df["y_prime"]))
```

```python
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(df["x"], df["y"])

print (slope, intercept, r_value, p_value, std_err)
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(df["x_prime"], df["y_prime"])

print (slope, intercept, r_value, p_value, std_err)
```
